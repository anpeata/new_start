```python
df = pd.read_excel(r'C:\Users\apkingpin\Downloads\Telegram Desktop\Teams.xlsx', index_col=0).reset_index()
# df['Tên NV'] = df['Tên NV'].str[-3:]
df['Tên NV'] = df['Tên NV'].str.split(' ').str[2]
df = df[df['Tên NV'].notna()]
df.drop('Phone',1).iloc[:,0:5]
```

```python
for index, row in df.iterrows():
    print(row[1], row[3])
```

```python
name = []
code = []

for index, row in df.iterrows():
    # print(row[1], row[2], row[3])
    l = []
    for letter in row[3].lower():
        if letter == 'a' or letter == 'k' or letter == 'u':
            l.append(4)
        elif letter == 'b' or letter == 'l' or letter == 'v':
            l.append(5)
        elif letter == 'c' or letter == 'm' or letter == 'w':
            l.append(6)
        elif letter == 'd' or letter == 'n' or letter == 'x':
            l.append(7)
        elif letter == 'e' or letter == 'o' or letter == 'y':
            l.append(8)
        elif letter == 'f' or letter == 'p' or letter == 'z':
            l.append(9)
        elif letter == 'g' or letter == 'q':
            l.append(0)
        elif letter == 'h' or letter == 'r':
            l.append(1)
        elif letter == 'i' or letter == 's':
            l.append(2)
        elif letter == 'j' or letter == 't':
            l.append(3)
        else:
            pass
        
    s = [str(i) for i in l]
    res = int("".join(s))
    # res = s
    # res = res / row[2] 
    # res = str('%.0f' % res)[10:15] # int(str(res)[:2])
    # print(row[1], ' ', res)
    
    res = '%.0f' % (res / row[2])
    name.append(row[1])
    code.append(res)
    
df_code = pd.DataFrame()
df_code['name'] = name
df_code['code'] = code
df_code
```

```python
df_code['0:5'] = df_code['code'].str[0:5]
df_code['5:10'] = df_code['code'].str[5:10]
df_code['10:15'] = df_code['code'].str[10:15]
df_code['15:20'] = df_code['code'].str[15:20]
df_code['20:25'] = df_code['code'].str[20:25]
```
